## According to GENDER

```python
ax = sns.countplot(x="Gender", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

The bar chart reveals a roughly equal purchase ratio between males and females.

```python
Total_sales_gen = df.groupby(["Gender"] ,as_index=False)["Total"].sum().sort_values(by = "Total", ascending=False)
sns.barplot(x ="Gender" , y ="Total", data=Total_sales_gen)


#Table Form
Total_sales_gen = df.groupby(["Gender"] ,as_index=False)["Total"].sum().sort_values(by = "Total", ascending=False)
Total_sales_gen
```

## According to Payment type

```python
ax=sns.countplot(data=df,x="Payment",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

From the above graph we can clearly see that the Ewallet, Cash Payment are approximately same and Credit card payment are less

## According to Product line

```python
plt.figure(figsize=(12, 8))
plt.xlabel("Product Line", fontsize=14)
plt.ylabel("Count", fontsize=14)
ax=sns.countplot(data=df,x="Product line")
for bars in ax.containers:
    ax.bar_label(bars)

```

From the Above graph we can easily see that Fashion accessories followed by Food and bevarages are in top where amount is most Spent

```python
plt.figure(figsize=(12, 8))
ax = df.groupby(["Product line"] ,as_index=False)["Customer Rating"].sum().sort_values(by = "Customer Rating", ascending=False)
sns.barplot(x ="Product line" , y ="Customer Rating", data=ax)
```

From the above graph we can clearly see that Fashion Accessories has highesht Customer Rating followed by Food and bevarages

```python
plt.figure(figsize=(12, 8))
ax = df.groupby(["Product line"] ,as_index=False)["Quantity"].sum().sort_values(by = "Quantity", ascending=False)
sns.barplot(x ="Product line" , y ="Quantity", data=ax)

```

According to Quantity we can see that Electronic accessories are in top, Therefore Electronic and accessories products are sold in more Quantity then Fashion accessories
